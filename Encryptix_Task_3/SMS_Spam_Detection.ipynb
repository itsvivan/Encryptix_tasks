```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/spam.csv',encoding='latin-1')
df.head()
```

```python
df.isnull().sum()
```

```python
df.tail()
```

```python
# removing all the columns that contains mostly all the Null Values and barely any Data
df1=df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
df1.head()
```

```python
le=LabelEncoder()
df1['spam']=le.fit_transform(df1['v1'])
df1.head()
```

```python
X=df1['v2']
y=df1['spam']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=3)
```

```python
v=TfidfVectorizer(min_df=1,stop_words='english')  # min_df ignores the words with the score less than 1   also stop words ignores the words that are commonlu used in  the english
X_train_features=v.fit_transform(X_train)
X_test_features=v.transform(X_test)
```

```python
print(v.vocabulary_)
```

```python
v.get_feature_names_out() # to get the vocabulary in the indexed order
```

```python
models={
    'LogisticRegression':LogisticRegression(),
    'SVC':SVC()
}
```

```python
k=list(models.values())
for i in range(len(list(models))):
  model=k[i]
  model.fit(X_train_features,y_train)

  y_train_pred=model.predict(X_train_features)
  y_test_pred=model.predict(X_test_features)

  train_acc=accuracy_score(y_train,y_train_pred)
  test_acc=accuracy_score(y_test,y_test_pred)

  print(list(models.keys())[i])
  print('Train Accuracy:',train_acc)
  print('Test Accuracy:',test_acc)


```

```python
#using SVC
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler  # For feature scaling

param_grid = {
    'C': [0.1, 1, 10, 100],  # Regularization parameter
    'kernel': ['linear', 'rbf'],  # Kernel function
    'gamma': [0.01, 0.1, 1],  # Kernel coefficient (for 'rbf' kernel)
}
```

```python
svc_classifier = SVC()
search_cv = GridSearchCV(svc_classifier, param_grid, cv=5)
```

```python
search_cv.fit(X_train_features, y_train)
```

```python
# Access best hyperparameters and model
best_params =search_cv.best_params_
best_model = search_cv.best_estimator_
best_score=search_cv.best_score_
print(best_score)
```

```python
# Print the best hyperparameters
print("Best Hyperparameters:", best_params)
```

```python
y_pred=best_model.predict(X_test_features)
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
```
